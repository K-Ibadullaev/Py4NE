```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500) # s
f1 = 5.0 # Hz
f2 = 8.0 # Hz
om1 = 2*np.pi*f1
om2 = 2*np.pi*f2
A1 = 0.8
A2 = 0.5
y = A1*np.sin(om1*t) + A2*np.sin(om2*t)

plt.figure()
plt.plot(t, y)
plt.xlabel('t in s')
plt.ylabel('Amplitude')
plt.grid(ls='-')
```
