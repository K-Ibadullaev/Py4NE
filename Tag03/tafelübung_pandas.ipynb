```python
plt.figure()
plt.plot(df1.Zeit*1e3, df1.Kraft/1e6)
plt.xlabel('Zeit in $\mu$s')
plt.ylabel('Kraft in kN')
plt.savefig('kraft.png', dpi=300)
```

```python
df1.plot(x='Zeit', y='Kraft')
plt.ylabel('Kraft in mN')
plt.grid(ls='--')
```

```python
fig, axes = plt.subplots(4, 1, figsize=(11, 12))

axes[0].plot(df1.Zeit, df1.Kraft)
for i in range(1, 4):
    axes[i].plot(df1.Zeit, df1['Y-Axis.%d' % i])
for i in range(4):
    axes[i].set(xlabel='Zeit in s', ylabel='Kraft in mN')
    axes[i].grid(ls='--')
fig.tight_layout()
```

```python
axes
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

axes[0, 0].plot(df1.Zeit, df1.Kraft)
ip = 1
for i in range(2):
    for j in range(2):
        if i == 0 and j == 0:
            continue
        axes[i, j].plot(df1.Zeit, df1['Y-Axis.%d' % ip])
        ip += 1
for i in range(2):
    for j in range(2):
        axes[i, j].set(xlabel='Zeit in s', ylabel='Kraft in mN')
        axes[i, j].grid(ls='--')
fig.tight_layout()
```
